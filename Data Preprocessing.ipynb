## Import Packages

```python
import pandas as pd
import numpy as np
import json
```

# Process Match Results

```python
with open('/content/drive/MyDrive/DSA Project/Dataset/Match Summary.json') as f:
    data=json.load(f)
data
```

```python
df_match = pd.DataFrame(data['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
df_match.drop(columns =['scorecard'], axis=1,inplace=True)
df_match.head()
```

```python
df_match.insert(6, 'match_id', range(1, 1 + len(df_match)))
```

```python
df_match.to_csv('/content/drive/MyDrive/DSA Project/CSV_Files/dim_match_summary.csv', index = False)
```

## Match ID Dictionary to link with other tables

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
```

# Batting Summary

```python
with open('/content/drive/MyDrive/DSA Project/Dataset/Batting_Summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['battingSummary'])
  
df_batting = pd.DataFrame(all_records)
df_batting.head(11)
```

```python
df_batting['out/not_out'] = df_batting.dismissal.astype(str).apply(lambda x: "out" if x!="nan"  else "not_out")
df_batting.head(22)
```

```python
df_batting['match_id'] = df_batting['match'].map(match_ids_dict)
df_batting.head(55)
```

```python
match_ids_dict
```

```python
df_batting.drop(columns=["dismissal"], inplace=True)
df_batting.head(10) 
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting.head()
```

```python
df_batting.shape
```

```python
df_batting.to_csv('/content/drive/MyDrive/DSA Project/CSV_Files/fact_bating_summary.csv', index = False)
```
