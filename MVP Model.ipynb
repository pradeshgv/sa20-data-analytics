## Data Preprocessing

```python
df_mvp.columns
```

```python
df_mvp=df_mvp[["Matches","Wickets","DotBalls","Fours","Sixes","IndexValue"]]
```

## Linear Regression

```python
lr=LinearRegression()
```

```python
X=df_mvp.drop(["IndexValue"],axis=1)
```

```python
y=df_mvp["IndexValue"]
```

```python
lr.fit(X,y)
```

```python
lr.coef_,lr.intercept_
```

```python
lr.score(X,y)
```

```python
lr.predict([[15	,8,	76,	49	,12]])
```
